```python
from scipy.optimize import root_scalar as rs

def f(x) :
    return x**3-4*x-9

root = rs(f, bracket=[2, 3], method='brentq')
print("Root:",root.root)
print("Root:",root)
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the ODE function
def ode45(t, y):
    return -2 * y

# Solve the ODE y' = -2y, y(0) = 1 over the interval [0, 5]
solution = solve_ivp(ode45, [0, 5], [1], t_eval=np.linspace(0, 5, 100))

# Plot the results
plt.plot(solution.t, solution.y[0], label="y(t)")
plt.title("Solution to ODE: y' = -2y, y(0) = 1")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.grid()
plt.show()

solution.t, solution.y[0]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a grid of x and y values
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

# Compute the Z values for the function z = x^2 + y^2
Z = X**2 + Y**2

# Plot the 3D surface
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k')

# Add labels and title
ax.set_title('3D Plot of z = x^2 + y^2')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Show the color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# Display the plot
plt.show()
```

```python
# Python code to calculate binary representations of fractions up to 6 places

import numpy as np

def binary_representation(fraction, precision=6):
    binary_str = "0."
    for _ in range(precision):
        fraction *= 2
        if fraction >= 1:
            binary_str += "1"
            fraction -= 1
        else:
            binary_str += "0"
    return binary_str

# Fractions to analyze
fractions = [1/3, 1/5, 1/6, 1/7, 1/9]

# Calculate binary representations
results = {}
for fraction in fractions:
    binary = binary_representation(fraction)
    results[fraction] = binary

results
```
